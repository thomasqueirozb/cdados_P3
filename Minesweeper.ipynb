```python
from random import randint

rows = 16
cols = 30
bombmax = 99
```

```python
class Cell:
    def __init__(self, i, j):
        self.i = i
        self.j = j
        self.neighborCount = 0

        self.bee = False
        self.revealed = False

    def countBees(self, arr):
        if self.bee:
            self.neighborCount = -1
            return

        total = 0
        for xoff in range(-1, 2, 1):
            i = self.i + xoff
            if i < 0 or i >= len(arr): continue

            for yoff in range(-1, 2, 1):
                j = self.j + yoff
                if j < 0 or j >= len(arr[0]): continue

                neighbor = arr[i][j]
                if neighbor.bee: total += 1

        self.neighborCount = total


    def reveal(self):
        self.revealed = True
        if self.neighborCount == 0:
            # flood fill time
            self.floodFill()

    def floodFill(self):
        global grid
        for xoff in range(-1, 2, 1):
            i = self.i + xoff
            if i < 0 or i >= rows: continue

            for yoff in range(-1, 2, 1):
                j = self.j + yoff
                if j < 0 or j >= cols: continue
                if not grid[i][j].revealed: grid[i][j].reveal()
```

```python
def constrain(num, lBound, uBound):
    if lBound <= num <= uBound:
        return num
    if num > uBound:
        return uBound
    return lBound
```

```python
def createMap(rows, cols, bombmax, clickX = -1, clickY = -1):
    total = cols * rows

    available = []
    arr = []

#   Criar uma lista bi-dimensional (arr) com as rows X cols com objetos Cell
#   Criar uma lista (available) com as possíveis posições de bombas

    for x in range(rows):
        arr.append([])
        for y in range(cols):
            arr[x].append(Cell(x, y))
            available.append([x, y])

#   Esse if roda apenas se o primeiro click do jogo não for uma célula vazia
    if clickX >= 0 and clickY >= 0:
        print('Clique')

#       Pega as células em volta da célula clicada
        nearCells = []
        for xoff in range(-1, 2, 1):
            i = clickX + xoff
            if i < 0 or i >= rows: continue

            for yoff in range(-1, 2, 1):
                j = clickY + yoff
                if j < 0 or j >= cols: continue
                nearCells.append([i, j])

#         print(f'{len(available)} - {len(nearCells)}',end =' = ')

        bombmax = constrain(bombmax, 0, total - len(nearCells))
        
        temp = []
        for i in available:
            if i not in nearCells:
                temp.append(i)
        available = temp
#                 print(el)
#                 available.pop(i)
#         print(len(available))


    bombmax = constrain(bombmax, 0, total)

    # Create all bombs
    bomb = 0

    while bomb < bombmax:
        index = randint(0, len(available) - 1)
        item = available.pop(index)
        if item == [0, 0]:print(item,available)
        arr[item[0]][item[1]].bee = True
        bomb += 1

    for x in range(len(arr)):
        for y in range(len(arr[x])):
            arr[x][y].countBees(arr)

    return arr
```
